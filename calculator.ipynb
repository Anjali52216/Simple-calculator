```python
from tkinter import *



def click(value):
    current=display.get()
    display.delete(0,END)
    display.insert(0,str(current)+str(value))

def evaluate():
    try:
        a= display.get()
        result=eval(a)
        display.delete(0,END)
        display.insert(0,str(result))
    except:
        display.delete(0,END)
        display.insert(0,'ERROR')

def clear():
    display.delete(0,END)
window = Tk()
window.title('simple calculator')
window.geometry('300x465')
display=Entry(window,width=400,font=('Arial',25))
display.place(x=0,y=0,width=300,height=90)
btn_a=Button(window,text='(',command =lambda:click('('))
btn_a.place(x=0,y=90,width=75,height=75)
btn_a=Button(window,text=')',command =lambda:click(')'))
btn_a.place(x=75,y=90,width=75,height=75)
btn_a=Button(window,text='%',command =lambda:click('%'))
btn_a.place(x=150,y=90,width=75,height=75)
btn_a=Button(window,text='AC',command =clear)
btn_a.place(x=225,y=90,width=75,height=75)
btn_a=Button(window,text='*',command =lambda:click('*'))
btn_a.place(x=0,y=165,width=75,height=75)
btn_a=Button(window,text='+',command =lambda:click('+'))
btn_a.place(x=75,y=165,width=75,height=75)
btn_a=Button(window,text='-',command =lambda:click('-'))
btn_a.place(x=150,y=165,width=75,height=75)
btn_a=Button(window,text='9',command =lambda:click('9'))
btn_a.place(x=225,y=165,width=75,height=75)
btn_a=Button(window,text='8',command =lambda:click('8'))
btn_a.place(x=0,y=240,width=75,height=75)
btn_a=Button(window,text='7',command =lambda:click('7'))
btn_a.place(x=75,y=240,width=75,height=75)
btn_a=Button(window,text='6',command =lambda:click('6'))
btn_a.place(x=150,y=240,width=75,height=75)
btn_a=Button(window,text='5',command =lambda:click('5'))
btn_a.place(x=225,y=240,width=75,height=75)
btn_a=Button(window,text='4',command =lambda:click('4'))
btn_a.place(x=0,y=315,width=75,height=75)
btn_a=Button(window,text='3',command =lambda:click('3'))
btn_a.place(x=75,y=315,width=75,height=75)
btn_a=Button(window,text='2',command =lambda:click('2'))
btn_a.place(x=150,y=315,width=75,height=75)
btn_a=Button(window,text='1',command =lambda:click('1'))
btn_a.place(x=225,y=315,width=75,height=75)
btn_a=Button(window,text='0',command =lambda:click('0'))
btn_a.place(x=0,y=390,width=75,height=75)
btn_a=Button(window,text='.',command =lambda:click('.'))
btn_a.place(x=75,y=390,width=75,height=75)
btn_a=Button(window,text='/',command =lambda:click('/'))
btn_a.place(x=150,y=390,width=75,height=75)
btn_a=Button(window,text='=',command =evaluate)
btn_a.place(x=225,y=390,width=75,height=75)
window.mainloop()
```
